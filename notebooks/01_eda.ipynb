## Dataset Overview

- Rows: (value from df.shape)
- Columns: (value from df.shape)
- Target: `Machine failure`
- Failure rate: X% (from value_counts(normalize=True))
- No missing values / some missing values in (columns) (based on df.isna().sum())

```python
numeric_cols = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]"
]

df[numeric_cols].hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Machine failure", y="Torque [Nm]")
plt.title("Torque vs Machine Failure")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Machine failure", y="Rotational speed [rpm]")
plt.title("Rotational Speed vs Machine Failure")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Machine failure", y="Tool wear [min]")
plt.title("Tool Wear vs Machine Failure")
plt.show()
```

```python
df["Type"].value_counts()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Type", hue="Machine failure")
plt.title("Machine Failure Count by Type")
plt.show()
```

## EDA Insights

- The dataset is imbalanced: only about X% of records are failures.
- Higher tool wear appears more common among failed machines.
- Certain machine types (e.g., Type H) show a higher proportion of failures compared to others.
- Torque and rotational speed distributions differ slightly between failure and non-failure cases.
